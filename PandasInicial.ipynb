## Reemplazo de espacios en los nombres de las columnas

Como pudiste observar no fue posible usar la notación punto(.) cuando el atributo tenía espacios en el nombre. Por eso es una buena práctica eliminar los espacios de los nombres de los atributos. 
Una forma de hacer este reemplazo es así:

```python
data.columns = data.columns.str.replace(' ', '_')
```

```python
data
```

# Filtrado de datos

```python
# Podemos tener los datos de una sola columna, por ejemplo el cloruro o chloride
data.chlorides
```

```python
data.chlorides.describe()
```

Usaremos el Chloride para consultar los que tienen concentraciones más altas
Ver artículo [del Cloruro en los vinos](https://www.scielo.br/j/cta/a/HQsrPrPMNZYgRzSKtrjHyHh/?lang=en)

```python
query1 = data[data.chlorides > 0.06].sort_values( 'alcohol' )
```

```python
query1
```

```python
query1.plot.scatter(x='chlorides', y='alcohol', figsize=(12,12))
```

```python
query1.plot.scatter(x='quality', y='chlorides', figsize=(12,12))
```
